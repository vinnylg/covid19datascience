```python
import pandas as pd
from bulletin.utils.normalize import normalize_text
from bulletin.utils.static import Municipios
```

```python
mun = Municipios().groupby(['uf','estado']).count().reset_index()
est = mun[['uf','estado']].copy()
est['estado'] = est['estado'].apply(normalize_text)
```

```python
rt = pd.read_excel('rt.xlsx')
rt.columns = ['uf','rt']

rt = pd.merge(est,rt,on='uf',how='left')
rt
```

```python
inferno = pd.read_excel('inferno.xlsx')
inferno.columns = [ x.lower() for x in inferno.columns ]
inferno['estado'] = inferno['estado'].apply(normalize_text)

inferno = pd.merge(inferno, rt, on='estado',how='left')
```

```python
inferno.sort_values('vacinas aplicadas').to_excel('rt_vacina.xlsx')
```
